```python
# calculating % of missing values for each column 

missing_percentage = df.isnull().mean()* 100
print(missing_percentage)
```

```python
#drop col where 70% data missing

col_to_drop = missing_percentage[missing_percentage > 70].index
df_cleaned = df.drop(columns = col_to_drop)
```

```python
print("orignal shape:", df.shape)
print("new shape", df_cleaned.shape)
```

```python
# filling missing values with median

df_final = df_cleaned.fillna(df_cleaned.median(numeric_only = True))
```

```python
print("missing values after imputation:",df_final.isnull().sum().sum())
```

```python
missing_cols = df_final.columns[df_final.isnull().any()]
print(df_final[missing_cols].isnull().sum())
```

```python
for col in missing_cols:
    if df_final[col].dtype == 'object':
        df_final[col] = df_final[col].fillna(df_final[col].mode()[0])
```

```python
print(df_final.isnull().sum().sum())
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
object_cols = df_final.select_dtypes(include = ['object']).columns
print(object_cols)
```

```python
le = LabelEncoder()
```

```python
df_final[col] = le.fit_transform(df_final[col])
```

```python
print(df_final.head())
```

```python
for col in object_cols:
    df_final[col] = df_final[col].astype('category').cat.codes
```

```python
print(df_final.head())
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df_final.drop(columns = ['diabetes_mellitus'])
y = df_final['diabetes_mellitus']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
model = RandomForestClassifier(n_estimators = 100,random_state = 42)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test,y_pred)
print("model accuracy:",accuracy)
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
importances = pd.Series(model.feature_importances_,index = x.columns)
importances.nlargest(10).plot(kind = 'barh')
plt.title("top 10 influential features")
plt.show()
```
